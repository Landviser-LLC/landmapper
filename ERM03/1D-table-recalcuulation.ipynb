```python
import csv
import math
import pandas as pd
```

```python
def calculate_k_multiply(AB_half:float, MN: float = 2)->float:
    return math.pi * (AB_half - MN / 2) * (AB_half + MN / 2) / MN
```

```python
def recalcule_ER(AB_half:float, k_set, r):
    return r * calculate_k_multiply(AB_half) / k_set 
```
